```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
learning_history = list(np.load('history_latest.npy', allow_pickle=True))

iters = [h["iter"] for h in learning_history]
desired_returns = [h["desired_return"] for h in learning_history]
desired_horizons = [h["desired_horizon"] for h in learning_history]
actual_returns = [h["actual_return"] for h in learning_history]
training_loss = [h["training_loss"] for h in learning_history]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 10))
axs[0, 0].plot(iters, desired_returns)
axs[0, 0].set_title('Desired return')
axs[0, 1].plot(iters, desired_horizons)
axs[0, 1].set_title('Desired horizon')
axs[1, 0].plot(iters, training_loss)
axs[1, 0].set_title('Training loss')
axs[1, 1].plot(iters, actual_returns)
axs[1, 1].set_title('Actual return')
plt.savefig('learning_history.png')
plt.show()
```

```python
plt.subplots(figsize=(16, 8))
plt.plot(iters, desired_returns)
plt.plot(iters, actual_returns)
plt.legend(['Desired return', 'Actual return'])
plt.savefig('desired_vs_actual.png')
plt.show()
```
